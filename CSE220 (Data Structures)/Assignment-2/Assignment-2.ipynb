```python
class myList:
    def __init__(self,a=None):
        self.head = Node(-1)
        self.size = 0

        if a != None:
            self.head.next = a.head.next
            length = self.head.next

            while length != None:
                self.size += 1
                length = length.next

    def create_list(self,key):
        a=self.head
        for x in key:
            self.size+=1
            b=Node(x)
            a.next=b
            a=a.next
            x+=1
```

```python
def insert(self, newElement):
        self.idx = None
        a = Node(newElement)
        b = self.head
        while b.next != None:
            b = b.next
            if b.element == newElement:
                print("Not empty")
        if self.idx == None:
            b.next = a
            self.size += 1
        elif self.size < self.idx:
            print("Invalid")
        else:
            b = self.head
            for x in range(self.idx):
                b = b.next
                x += 1
            a.next = b.next
            b.next = a
            self.size += 1
```

Post-condition: Removes the element from a list that contains the deleteKey and returns the deleted key value.


```python
def remove(self, deletekey):
        b = self.head
        while b.next != None:
            if b.next.element == deletekey:
                i = b.next
                b.next = i.next
                i.next = None
                self.size -= 1
                return i.element
            b = b.next
        return None
```
