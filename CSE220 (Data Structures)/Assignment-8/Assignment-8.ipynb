```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
def height(root):
    if root is None:
	    return 0
    else:
	    left = height(root.left)
	    right = height(root.right)
    if left > right:
	    return left+1
    else:
	    return right+1
def get_level(root,value,level):
    if root == None:
	    return 0
    if root.value == value:
	    return level
    low_level = get_level(root.left,value,level+1)
    if low_level != 0:
	    return low_level
    low_level = get_level(root.right, value, level+1)
    return low_level
def check(root,value):
    return get_level(root,value,1)
def Preorder(root):
    if root:
	    print(root.value)
	    Preorder(root.left)
	    Preorder(root.right)
def Inorder(root):
    if root:
        Inorder(root.left)
        print(root.value)
        Inorder(root.right)
def Postorder(root):
    if root:
	    Postorder(root.left)
        Postorder(root.right)
        print(root.value)
def same_tree_checker(tree1,tree2):
    if tree1 is None and tree2 is None:
	    return True
    if tree1 is not None and tree2 is not None:
	    return tree1.value == tree2.value and same_tree_checker(tree1.left,tree2.left) and same_tree_checker(tree1.right,tree2.right)
	return False
def copy(root):
    if root.value != None:
	    new_root = Node(root.value)
    if root.left != None:
	    new_root.left = root.left
    if root.right != None:
	    new_root.right = root.right
    print(new_root.value)
    print(new_root.left.value)
    print(new_root.right.value)
    print(new_root.left.left.value)
    print(new_root.left.right.value)
#===============================Tester===============================#
# test code: 1
print("===========================================")
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
print("The height of the tree is:",height(root))
# test code: 2
print("===========================================")
print("The level of valus:")
for x in range(1,6):
    level = check(root,x)
    if level:
	    print("level of",x,"is",check(root,x))
    else:
	    print(x,"not present")
# test code: 3
print("===========================================")
print("Preorder:")
Preorder(root)
# test code: 4
print("===========================================")
print("Inorder:")
Inorder(root)
# test code: 5
print("===========================================")
print("Postorder:")
Postorder(root)
# test code: 6
print("===========================================")
root2 = Node(1)
root2.left = Node(2)
root2.right = Node(3)
root2.left.left = Node(4)
root2.left.right = Node(5)
if same_tree_checker(root,root2):
    print("Similar Tree Checker: Trees are same")
else:
    print("Similar Tree Checker: Trees are different")
# test code: 7
print("===========================================")
print("The New Tree:")
copy(root)
```
